```python
from bs4 import BeautifulSoup
import requests
```

```python
url = input('Enter a url: ')
```

```python
r = requests.get('http://' +url)
```

```python
data = r.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
print(soup.get_text)
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
for image in soup.find_all('img'):
    print(image.get('src'))
```
